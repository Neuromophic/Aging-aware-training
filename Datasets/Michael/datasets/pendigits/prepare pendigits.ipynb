```python
def get_pendigits(test=False):
    ''' prepares the pendigits dataset 

        test : bool : Return test data, else train data
    '''

    if test:
        s = 'test'
    else:
        s = 'train'

    # path to file
    path = os.path.join(os.getcwd(), 'pendigits_{}_original.csv'.format(s))
    tmp = pd.read_csv(path, sep=',', header=None)
    #tmp.columns = ['X_{}'.format(i) for i in range(16)] + ['Y']

    # normalize data
    x, y = torch.from_numpy(
        tmp.values[:, :16]/100.), torch.from_numpy(tmp.values[:, 16]).long()

    return x, y
```

```python
X_train, y_train = get_pendigits()
X_test, y_test = get_pendigits(test=True)
```

```python
print(f'Number of input features: {X_train.shape[1]}.')
print(f'Number of output classes: {torch.max(y_train)}.')
print(f'Number of training examples: {X_train.shape[0]}.')
print(f'Number of test examples: {X_test.shape[0]}.')
```

```python
X = torch.cat((X_train, X_test))
y = torch.cat((y_train, y_test))
```

```python
import pickle
Data = {'X': X, 'y': y}
with open('Dataset_Pendigits.p', 'wb') as f:
    pickle.dump(Data, f)
```
