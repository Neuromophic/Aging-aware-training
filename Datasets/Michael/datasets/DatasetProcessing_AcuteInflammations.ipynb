```python
import pandas as pd
```

```python
data = pd.read_csv('Acute Inflammations.csv', sep=';')
data.head()
```

```python
x_df = data.loc[:, 'X_1':'X_6']
y_df = data['Y']
```

```python
import torch
X, y = torch.from_numpy(x_df.values), torch.from_numpy(y_df.values).long()
X.shape, y.shape
```

```python
import pickle
Data = {'X': X, 'y': y}
with open('Dataset_AcuteInflammations.p', 'wb') as f:
    pickle.dump(Data, f)
```

```python
with open('Dataset_AcuteInflammations.p', 'rb') as f:
    Data_read = pickle.load(f)
Data_read
```
