```python
import os
import numpy as np
import torch
import pandas as pd
```

```python
data = pd.read_csv('arrhythmia.data', header=None)
data.head()
```

```python
# remove NaN
data.dropna(axis='columns', how='all', inplace=True)
data.dropna(axis='index', how='any', inplace=True)
# remove '?'
data = data.loc[:,(data!='?').all(axis=0)]
data = data.loc[(data!='?').all(axis=1),:]
data.head()
```

```python
X_full = data.loc[:,:278].astype('float')
y = data.loc[:,279].astype('float')
X_full.shape, y.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=12)
X = pca.fit_transform(X_full)
X.shape
```

```python
y = y.to_numpy()
unique = np.unique(y)
for i in range(len(unique)):
    y[y==unique[i]] = i
```

```python
X, y = torch.from_numpy(X), torch.from_numpy(y).long()
X.shape, y.shape
```

```python
import pickle
Data = {'X': X, 'y': y}
with open('Dataset_Arrhythmia.p', 'wb') as f:
    pickle.dump(Data, f)
```
