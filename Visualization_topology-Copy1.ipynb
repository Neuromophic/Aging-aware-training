```python
import os
import matplotlib.pyplot as plt
import numpy as np
import config
import sys
sys.path.append(os.path.join(os.getcwd(), 'Aging_Model'))
import FigureConfig as FC
```

```python
x = np.linspace(0,np.pi/4,1000)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x, y, '--', label='random guess', color=FC.Black);
plt.fill_between(x, y, z, alpha=0.3, label='PNN std', color=FC.Cyan);


plt.plot(x, y, label='PNN mean', color=FC.Cyan);
plt.fill_between(x, y, z, alpha=0.3, label='AAPNN std', color=FC.Pink);
plt.plot(x, z, label='AAPNN mean', color=FC.Pink);


plt.legend()
plt.show()
```
